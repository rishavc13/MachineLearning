```python
# import libraries
from sklearn import tree
from sklearn.datasets import load_iris
import numpy as np

# loading irisidata set
iris=load_iris()

x=[0,50,100]

# removing exactly one data from each flower for testing later
train_data=np.delete(iris.data,x,axis=0)

# removing the same from target
train_target=np.delete(iris.target,x)

# saving data to be used for testing
test_data=iris.data[x]
test_target=iris.target[x]

# calling classifier
clf=tree.DecisionTreeClassifier()

trained=clf.fit(train_data,train_target)

# making prediction
output=trained.predict([[9.2,5.6,3.6,7.1]])
print(output)
```

```python
# making graph of decision tree classifier
import graphviz
out_data=tree.export_graphviz(clf, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True, special_characters=True)
graphviz.Source(out_data)
```
